Ordered Pathway

```python
#Total well step length 5300H 3400V - NUNC 12x8 well plate
#Ordered pathway - Rows - columns 
rows = 8
cols = 12
hstep = 5275/(cols-1)
vstep = 3350/(rows-1)
for i in range(0,rows):
    for j in range(0,cols-1):
        move("Horz","Away",hstep)
        time.sleep(3)
    move("Horz","Towards",5275)
    move("Vert","Away",vstep)
move("Vert","Towards",3350+hstep)

```

Snaking Pathway

```python
#Total well step length 5300H 3400V - NUNC 12x8 well plate
#Snaking pathway - Rows - columns 
rows = 8
cols = 12
hstep = 5275/(cols-1)
vstep = 3350/(rows-1)
for i in range(0,rows-1,2):
    for j in range(0,cols-1):
        move("Horz","Away",hstep)
        time.sleep(3)
    move("Vert","Away",vstep)
    for j in range(0,cols-1):
        move("Horz","Away",hstep)
        time.sleep(3)
move("Vert","Towards",3350)
```
